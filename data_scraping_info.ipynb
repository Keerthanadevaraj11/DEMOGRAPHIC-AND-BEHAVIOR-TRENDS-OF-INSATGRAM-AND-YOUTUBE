```python
def get_post_info(data,profile_details,post_details):
    L = instaloader.Instaloader()
    for username in data:
        print(username)
        profile = instaloader.Profile.from_username(L.context, username)
        pd=[profile.username,profile.userid,profile.mediacount,profile.followers,profile.followees,profile.biography,profile.external_url]
        profile_details.loc[len(profile_details)]=pd
        posts=profile.get_posts()
        print("yes")
        c=0
        for post in posts:
            print(c)
            temp=[post.mediaid, post.profile, post.caption, post.date, post.location,  post.typename, post.mediacount, post.caption_hashtags, post.caption_mentions, post.tagged_users, post.likes, post.comments,  post.url,post.shortcode]
            post_details.loc[len(post_details)]=temp
            if c==50:
                break
            c+=1
    return profile_details,post_details
```
