```python
import pandas as pd
from ast import literal_eval
import networkx as nx
from pyvis.network import Network
```

```python
post=pd.read_csv("post_details.csv",converters={"caption hashtags":literal_eval,"caption mentions":literal_eval,"tagged users":literal_eval})
post
```

```python
source_target=[]
for i in post.index:
    for j in post["caption hashtags"][i]:
        source_target.append([post["username"][i],j])
        
g=nx.Graph()
for e in source_target:
    g.add_edge(e[0], e[1])
    
net=Network(height='750px', width='70%', bgcolor='#222222', font_color='white')
net.barnes_hut()

degree=nx.degree(g)

for e in source_target:
    src=e[0]
    dst=e[1]
    
    net.add_node(src, src, title=src)
    net.add_node(dst, dst, title=dst)
    net.add_edge(src, dst)
    
bet_cen1=nx.betweenness_centrality(g, weight='weight')
close_cen1=nx.closeness_centrality(g)
deg_cen1=nx.degree_centrality(g)
degree_centrality=sorted(deg_cen1.items(), key=lambda x:x[1],reverse=True)[:20]
closeness_centrality=sorted(close_cen1.items(), key=lambda x:x[1],reverse=True)[:20]
betweenness_centrality=sorted(bet_cen1.items(), key=lambda x:x[1],reverse=True)[:20]
for node in net.nodes:
    node['title']+=' <br>Betweenness Centrality: '+ \
    '{0:.4f}'.format(bet_cen1[node['id']]) + '<br>Degree Centrality: '+ '{0:.4f}'.format(deg_cen1[node['id']])+\
     '<br> Closeness Centrality: '+ '{0:.4f}'.format(close_cen1[node['id']])
    node['value']=100*degree[node['id']]
net.show_buttons()
net.show("hash.html")
net.save_graph("hash.html")
```

```python
u=post["username"].unique()
c=0
l=[degree_centrality,closeness_centrality,betweenness_centrality]
v=["degree_centrality","closeness_centrality","betweenness_centrality"]
for j,p in zip(l,v):
    print(f"\n{p}\n")
    for i in j:
        if not i[0] in u:
            c+=1
            print(i)
        if c==10:
            c=0
            break
```
